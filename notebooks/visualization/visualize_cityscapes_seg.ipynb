```python
import numpy as np
from PIL import Image
import cv2

import sys
sys.path.append('../../src/models')
from predict import SegmentationModel

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
model_path, scale = '../../models/cityscapes_v3_clahe/model_epoch_198', 0.5
mean_path = '../../data/cityscapes/mean.npy'
mean = np.load(mean_path)

model = SegmentationModel(model_path, mean, scale=scale)
```

```python
def visualize(image, mask, score, figsize=(12, 12)):
    fig = plt.figure(figsize=figsize)
    plt.imshow(mask)
    plt.show()
    
    overlay = (image * 0.5 + mask * 0.5).astype(np.uint8)
    fig = plt.figure(figsize=figsize)
    plt.imshow(overlay)
    plt.show()
    
    fig = plt.figure(figsize=figsize)
    plt.imshow(image)
    plt.show()

def apply_segmentation(image_path):
    pil_image = Image.open(image_path)
    image = np.asarray(pil_image, dtype=np.uint8)
    
    score, mask = model.apply_segmentation(pil_image)
    visualize(image, mask, score)
```

```python
path = '../../data/cityscapes/leftImg8bit_trainvaltest/leftImg8bit/test/munich/munich_000251_000019_leftImg8bit.png'
apply_segmentation(path)
```
